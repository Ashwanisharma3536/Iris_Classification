# To Select the Best performing Model and its hyperparameters using Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Sabhi Models import kiye
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# 2. Models aur unke GridSearchCV ke Hyperparameters ki Dictionary banayi
# Note: Parameter keys me 'model_' lagana zaroori hai kyunki pipeline me step ka naam 'model' hai
models_and_params = {
    'logis_reg': {
        'model': LogisticRegression(max_iter=1000),
        'params': {'model_C': [0.1, 1, 10]}
    },
    'rand_for': {
        'model': RandomForestClassifier(random_state=42),
        'params': {'model_n_estimators': [50, 100], 'model_max_depth': [5, None]}
    },
    'knn': {
        'model': KNeighborsClassifier(),
        'params': {'model_n_neighbors': [3, 5, 7]}
    },
    'deci_tree': {
        'model': DecisionTreeClassifier(),
        'params': {'model_max_depth': [3, 5, 10]}
    },
    'svc': {
        'model': SVC(),
        'params': {'model_C': [1, 10], 'model_kernel': ['rbf', 'linear']}
    },
    'ada_boost': {
        'model': AdaBoostClassifier(),
        'params': {'model_n_estimators': [50, 100]}
    },
    'vot_class': {
        'model': VotingClassifier(estimators=[
            ('lr', LogisticRegression(max_iter=1000)),
            ('rf', RandomForestClassifier(random_state=42)),
            ('knn', KNeighborsClassifier())
        ]),
        'params': {'model_voting': ['hard', 'soft']} # Voting classifier ke params
    }
}

# Natijo ko store karne ke liye dictionaries
best_estimators = {}
model_accuracies = {}

print("--- Training Models with GridSearchCV & Pipeline ---")

# 3. LOOP: Jo har ek model par pipeline aur gridsearch lagayega
for model_name, mp in models_and_params.items():
    # Base Pipeline banayi (Correct Tuple Format)
    base_pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('model', mp['model']) # Yahan model dynamically badlega
    ])

    # GridSearchCV apply kiya
    grid = GridSearchCV(base_pipeline, mp['params'], scoring='accuracy', cv=3, n_jobs=-1)
    grid.fit(x_train, y_train)

    # Best trained pipeline/model ko store kiya
    best_estimators[model_name] = grid.best_estimator_

    # Test data par predict karke actual accuracy nikali
    y_pred = grid.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    model_accuracies[model_name] = acc

    print(f"{model_name} trained. Test Accuracy: {acc*100:.2f}%")

# 4. Best Model Select Karna (Jis ki Accuracy sabse ZYADA ho)
best_model_name = max(model_accuracies, key=model_accuracies.get)
print(f"\n WINNER MODEL: {best_model_name} with {model_accuracies[best_model_name]*100:.2f}% Accuracy")

# 5. Winner Model ki Classification Report print karna
winner_pipeline = best_estimators[best_model_name]
winner_preds = winner_pipeline.predict(x_test)

print(f"\n--- Classification Report for {best_model_name} ---")
print(classification_report(y_test, winner_preds))

print("--- Confusion Matrix ---")
print(confusion_matrix(y_test, winner_preds))

# 6. Sabhi Models ki Performance ka Graph (Bar Plot) banana
plt.figure(figsize=(10, 5))
plt.bar(model_accuracies.keys(), model_accuracies.values(), color='skyblue', edgecolor='black')
plt.ylabel('Test Accuracy Score')
plt.title('Comparison of Model Accuracies')
plt.ylim(0, 1.0) # Accuracy 0 se 1 ke beech dikhane ke liye
for i, v in enumerate(model_accuracies.values()):
    plt.text(i, v + 0.02, f"{v*100:.1f}%", ha='center', fontweight='bold')
plt.show()
```

## Quick Tip
Whenever a confusion matrix has values only on the **diagonal** (10, 17, 11 in this case) and all other positions contain **0**,
it means the model has made **no classification errors** and achieved **100% accuracy**.

# 3. Classification Report
The Classification Report provides a detailed evaluation of the model's performance for each class.

### Support

Support indicates the number of actual samples belonging to each class in the test dataset:

* Class 0: 10 samples
* Class 1: 17 samples
* Class 2: 11 samples

Total samples = 38.

### Precision (1.00)

Precision measures how many of the predictions made for a class were actually correct. A precision score of **1.00** means that
every flower predicted as a particular class truly belonged to that class. No false positives were made.

### Recall (1.00)

Recall measures how many of the actual flowers belonging to a class were successfully identified by the model.
A recall score of **1.00** means the model found every flower of that class. No samples were missed.

### F1-Score (1.00)

The F1-Score is the harmonic mean of Precision and Recall. Since both Precision and Recall are perfect (1.00),
the F1-Score is also **1.00**, indicating excellent model performance.

## Important Note: Overfitting Check

* In Machine Learning, achieving **100% accuracy** is impressive, but it is also a reason to be cautious.

* A perfect score can sometimes indicate **Overfitting** or **Data Leakage**, where the model has learned the training data too well or
  where information from the training set accidentally appears in the test set.

* However, since you are working with the **Iris Dataset**, achieving 100% accuracy is quite common because it is a small, clean,
  and relatively simple dataset.

* If this were a real-world project with a larger and more complex dataset, you would need to verify that
  there was no overlap between the training and testing data and ensure that the model generalizes well to unseen data.
